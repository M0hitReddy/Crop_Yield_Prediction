```python
df_encoded=pd.get_dummies(df,columns=['State_Name','Crop','Crop_Type'])
del df_encoded['Yield_ton_per_hec']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=df_encoded.drop('Production_in_tons',axis=1)
y=df_encoded['Production_in_tons']
x.columns
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
regressor = RandomForestRegressor(n_estimators=65, random_state=42)
```

```python
regressor.fit(x_train,y_train)
```

```python
y_pred=regressor.predict(x_test)
```

```python
pd.DataFrame(y_pred).head(10)
```

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
```

```python
r2_score(y_test,y_pred)
```

```python
mse=mean_squared_error(y_test,y_pred)
```

```python
mean_absolute_error(y_test,y_pred)
```

```python
np.sqrt(mse)
```
